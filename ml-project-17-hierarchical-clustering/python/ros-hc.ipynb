# Hierarchical Clustering

## import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## import dataset

```python
dataset = pd.read_csv('../Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

## Use dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

## Training Hierarchical Clustering model on dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
print(y_hc)
```

## Visualising clusters

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c='red', label='1. High Income-Spend Least (Target)')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c='blue', label='2. Average Income-Spend Average (Low Prioritise)')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c='green', label='3. High Income-Spend Lots (Prioritise)')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s=100, c='cyan', label='4. Low Income-Spend Lots (Avoid)')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s=100, c='magenta', label='5. Low Income-Spend Least (Ignore)')

plt.title('Clusters of Customers')
plt.xlabel('Annual Income [k$]')
plt.ylabel('Spending Score [1-100]')
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```

* 1. try target cluster 1 to attract them more
* 2. Customers with Low income spend Least ignore cluster 2
* 3. Average spend and income, don't do much
* 4. Avoid targeting cluster 4 because low income are spending too much money
* 5. Don't do anything with cluster 5
